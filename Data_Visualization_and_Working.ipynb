```python
label_encoding_columns = ['verbatimScientificName', 'verbatimTraitName', 'traitName', 'traitID', 'traitUnit', 'taxonID']

numerical_columns = ['traitValue', 'measurementID', 'occurrenceID']
```

```python
from sklearn import preprocessing as prepro
label_encoder = prepro.LabelEncoder()
one_hot_encoder = prepro.OneHotEncoder()
for col in label_encoding_columns:
    data[col] = label_encoder.fit_transform(data[col])

z_score = prepro.StandardScaler()
for col in numerical_columns:
    data[col] = z_score.fit_transform(data[col].values.reshape(-1,1))

data
```

```python
dataframe_input = data.drop(columns=['traitID'])
dataframe_output = data['traitID']
```

```python
# Multilayer perceptron
from sklearn.neural_network import MLPClassifier


mlp_model = MLPClassifier()

input_train, input_test, output_train, output_test = train_test_split(dataframe_input, dataframe_output, test_size=0.3, random_state=100)

mlp_model.fit(input_train, output_train)
output_predict = mlp_model.predict(input_test)
print(classification_report(output_test, output_predict))
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

knn_model = KNeighborsClassifier()

input_train, input_test, output_train, output_test = train_test_split(dataframe_input, dataframe_output, test_size=0.3, random_state=100)

knn_model.fit(input_train, output_train)
output_pred = knn_model.predict(input_test)
print(classification_report(output_test, output_pred))
```

```python
# XG Boost
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

xg_model = GradientBoostingClassifier()
input_train, input_test, output_train, output_test = train_test_split(dataframe_input, dataframe_output, test_size=0.3, random_state=100)
xg_model.fit(input_train, output_train)
output_pred = xg_model.predict(input_test)
print(classification_report(output_test, output_pred))
```

```python
from sklearn.naive_bayes import GaussianNB

naive_bayes_model = GaussianNB()

input_train, input_test, output_train, output_test = train_test_split(dataframe_input, dataframe_output, test_size=0.3, random_state=100)

naive_bayes_model.fit(input_train, output_train)
output_pred = naive_bayes_model.predict(input_test)

print(classification_report(output_test, output_pred))
```

```python
# Artificial Neural Network
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense

input_train, input_test, output_train, output_test = train_test_split(dataframe_input, dataframe_output, test_size=0.3, random_state=100)

model = Sequential()
model.add(Dense(16,input_dim=8,activation='relu'))
model.add(Dense(12,activation='relu'))
model.add(Dense(1,activation='softmax'))
model.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy'])
history = model.fit(input_train, output_train, epochs=100, batch_size=64)
output_pred = model.predict(input_test)
print(classification_report(output_test, output_pred))
```

```python
# Support Vector Machine
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split

input_train, input_test, output_train, output_test = train_test_split(dataframe_input, dataframe_output, test_size=0.3, random_state=100)

svm_model = SVC()
svm_model.fit(input_train, output_train)
svm_output_pred = svm_model.predict(input_test)
print(classification_report(output_test, svm_output_pred))
```

```python
from sklearn.ensemble import StackingClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import train_test_split

base_estimator_1 = MLPClassifier()
base_estimator_2 = GradientBoostingClassifier()
base_final_estimator = [('mlp', base_estimator_1), ('xg', base_estimator_2)]

stacker_model = StackingClassifier(estimators = base_final_estimator)
input_train, input_test, output_train, output_test = train_test_split(dataframe_input, dataframe_output, test_size=0.3, random_state=100)
stacker_model.fit(input_train, output_train)
stacker_output_pred = stacker_model.predict(input_test)


print(classification_report(output_test, stacker_output_pred))
```
